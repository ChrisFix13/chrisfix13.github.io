```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
df = pd.read_csv('tweets.csv')
df
```

```python
vectorizer = CountVectorizer(max_features = 1000, analyzer='word', ngram_range=(1, 2))
Txt_rep_model = vectorizer.fit(df['text'])
full_tweets = vectorizer.transform(df['text'])
vectorizer.get_feature_names_out()
```

```python
full_tweets.toarray()
```

```python
query1 = vectorizer.transform(['americanair is very enjoyable']).toarray()
query2 = vectorizer.transform(['my flight was delayed']).toarray()
query3 = vectorizer.transform(['attendants are rude']).toarray()
```

## Looking for the Euclidean distance between vectors

```python
dist1 = np.linalg.norm(query1 - query2)
dist2 = np.linalg.norm(query2 - query3)
dist3 = np.linalg.norm(query1 - query3)
print(dist1)
print(dist2)
print(dist3)
```

```python
cos1 = cosine_similarity(query1, query2)
cos2 = cosine_similarity(query2, query3)
cos3 = cosine_similarity(query3, query1)
```

```python
print(cos1)
print(cos2)
print(cos3)
```
