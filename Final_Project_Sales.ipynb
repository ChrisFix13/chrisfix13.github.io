Looking at the first publisher, and comparing it with other publishers, we can see that at the end of the publisher's name, it's written "JP: Koei TecmoNA/PAL: Nintendo", which does coincide with other publishers. The next step is to change the name of that Publisher.

```python
df['publisher'] = df['publisher'].str.replace(pat = '.mw-parser-output .plainlist ol,.mw-parser-output .plainlist ul{line-height:inherit;list-style:none;margin:0;padding:0}.mw-parser-output .plainlist ol li,.mw-parser-output .plainlist ul li{margin-bottom:0}JP: Koei TecmoNA/PAL: Nintendo'
                                              , repl = 'JP: Koei TecmoNA/PAL: Nintendo')
display(df['publisher'].unique())

# Here we used .str.replace() to be able to target every string that contained
# the publisher we wanted to change.
```

```python
df_P_S = df[['publisher','copies_sold']].groupby('publisher').sum()
display(df_P_S)
```

Since we are focusing more on the genres, and publishers in general, we will combine the publishers that are divided into Japanese and the North American into one single publisher. And will remove the JP or NA seal and will only leave the name of the publishers.

```python
df['publisher'] = df['publisher'].str.replace(pat = 'JP: AtlusNA: SegaPAL: Nintendo', repl = 'Atlus')
df['publisher'] = df['publisher'].str.replace(pat = 'JP: ImagineerNA/PAL: Nintendo', repl = 'Imagineer')
df['publisher'] = df['publisher'].str.replace(pat = 'JP: Koei TecmoNA/PAL: Nintendo', repl = 'Koei Tecmo')
df['publisher'] = df['publisher'].str.replace(pat = 'JP: Konami', repl = 'Konami')
df['publisher'] = df['publisher'].str.replace(pat = 'JP: Square EnixNA/PAL: Nintendo', repl = 'Square Enix')
df['publisher'] = df['publisher'].str.replace(pat = 'JP: The Pokémon CompanyNA/PAL: Nintendo', repl = 'The Pokémon Company')
df['publisher'] = df['publisher'].str.replace(pat = 'JP: Xbox Game StudiosNA/PAL: Mojang', repl = 'Xbox Game Studios')
df['publisher'] = df['publisher'].str.replace(pat = 'NA/PAL: UbisoftJP/KOR: Nintendo[18]', repl = 'Ubisoft')
df['publisher'] = df['publisher'].str.replace(pat = 'The Pokémon CompanyNintendo', repl = 'The Pokémon Company')

display(df['publisher'].unique())
```

## Q3. Could you show us the performance of every Nintendo Switch game per year of release in order of highest copies sold to lowest?

#### Now I want to create a new column in the main Dataset that divides the games into their specific years of release.

```python
df
```

```python
# We define a function 'dates(x)' that will separate the videogames into 
# their specific year of release.

def dates(x):
    if x < '2018-01-01':
        return 2017
    elif '2018-01-01' <= x < '2019-01-01':
        return 2018
    elif '2019-01-01' <= x < '2020-01-01':
        return 2019
    elif '2020-01-01' <= x < '2021-01-01':
        return 2020
    elif '2021-01-01' <= x < '2022-01-01':
        return 2021
    else:
        return 2022
```

```python
df['dates'] = df['release_date'].apply(dates)
df.head(10)

# We use .apply() to be able to plug in the function 'dates()' into the dataset
# and that will help us create the new column.
```
